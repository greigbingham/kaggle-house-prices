### Data Manipulation

#### provide categories to gain better correlations

```python
house_prices_train_transform = house_prices_train.copy()

for x in all_cat_features:
    house_prices_train_transform[x] = house_prices_train_transform[x].astype('category').cat.codes

house_prices_train_transform.head()
```

#### Set correlations

```python
corr_with_sale_price = house_prices_train_transform[house_prices_train_transform.columns[1:]].corr()['SalePrice'][:-1]
corr_with_sale_price
```

```python
corr_with_sale_price.to_csv("Working/corr_with_sale_price.csv")
```

```python
corr_all = house_prices_train_transform.corr().abs()

s = corr_all.unstack()
so = s.sort_values(kind="quicksort")

so.to_csv("Working/all_corr.csv")
```

```python
import numpy as np

house_prices_train_transform.fillna(0, inplace = True)

for i in house_prices_train_transform.columns:
    house_prices_train_transform[i] = np.abs(house_prices_train_transform[i])

house_prices_train_transform.to_csv("Working\data_no_nan_no_blank.csv")

house_prices_train_transform.shape
```

### Univariate selection

```python
#import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#X = house_prices_train_transform.iloc[:,0:80]  #independent columns
X = house_prices_train_transform.drop("SalePrice", axis=1)
y = house_prices_train_transform["SalePrice"]   #target column i.e price

#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

```python
featureScores.to_csv("Working/featurescores.csv")
```

### Feature Importance

```python

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

### Correlation with heat map

```python
import seaborn as sns

#get correlations of each features in dataset
corrmat = house_prices_train_transform.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(house_prices_train_transform[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

## OK features model...

```python
# want the most impoartant columns!

most_important = featureScores.nlargest(10,'Score')

X = pd.DataFrame()

for i in most_important['Specs']:
    X[i] = house_prices_train_transform[i]


X
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

np.random.seed(17)

#X Defined in previous cell
y = house_prices_train_transform["SalePrice"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
actual_test_set = pd.read_csv("Data/test.csv")
```

```python
for x in all_cat_features:
    actual_test_set[x] = actual_test_set[x].astype('category').cat.codes
```

```python
actual_test_set.fillna(0, inplace = True)

for i in actual_test_set.columns:
    actual_test_set[i] = np.abs(actual_test_set[i])
```

```python
test_set = pd.DataFrame()

for i in most_important['Specs']:
    test_set[i] = actual_test_set[i]

test_set
```

```python
actual_preds = model.predict(test_set)
actual_preds
```

```python
output_df = pd.DataFrame()

output_df["Id"] = list(range(1461,2920))
output_df["SalePrice"] = actual_preds

output_df.head()
```

```python
output_df.to_csv('Submissions\submission_20220302.csv', index=False)
```

## Recursive Feature Extraction

```python
from sklearn.feature_selection import RFE

np.random.seed(17)

model = LinearRegression()

X = house_prices_train_transform.drop("SalePrice", axis=1)
y = house_prices_train_transform["SalePrice"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

rfe = RFE(model, n_features_to_select=10, step=2)
fit = rfe.fit(X_train, y_train)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
fit.score(X_test, y_test)
```

```python
actual_preds = fit.predict(actual_test_set)

output_df = pd.DataFrame()

output_df["Id"] = list(range(1461,2920))
output_df["SalePrice"] = actual_preds

output_df.head()

output_df.to_csv('Submissions\submission_20220303_recursive.csv', index=False)

output_df.head()
```

## PCA

```python
from sklearn.decomposition import PCA
# load data

np.random.seed(17)

X = house_prices_train_transform.drop("SalePrice", axis=1)
y = house_prices_train_transform["SalePrice"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

pca = PCA(n_components=3)
fit = pca.fit(X_train)

# summarize components
print("Explained Variance: %s" % fit.explained_variance_ratio_)
print(fit.components_)
```

### Extra trees classifier

```python
from sklearn.ensemble import ExtraTreesClassifier

# feature extraction
model = ExtraTreesClassifier(n_estimators=10)
model.fit(X_train, y_train)
print(model.feature_importances_)
```

### More RFE

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVR

np.random.seed(17)

X = house_prices_train_transform.drop("SalePrice", axis=1)
y = house_prices_train_transform["SalePrice"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

estimator = SVR(kernel="linear")
selector = RFE(estimator, n_features_to_select=5, step=1)
selector = selector.fit(X_train, y_train)
selector.support_
```

```python
actual_preds = selector.predict(actual_test_set)

output_df = pd.DataFrame()

output_df["Id"] = list(range(1461,2920))
output_df["SalePrice"] = actual_preds

output_df.head()

output_df.to_csv('Submissions\submission_20220303_RFE_SVR.csv', index=False)

output_df.head()
```
